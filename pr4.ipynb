```python
with open('input/4') as f:
    dat = f.read()
```

```python
import regex as re

datP = re.findall('\[(\d{4})-(\d{2})-(\d{2}) (\d{2}):(\d{2})] \w* (\S*)',dat)
print(datP[0:15])
```

```python
# create timestamp, event tuple


sorted_list = sorted([(int(i[0]+i[1]+i[2]+i[3]+i[4]),i[5]) for i in datP], key=lambda x: x[0])

for l in sorted_list:
    print(l)
```

```python
import numpy as np

```

```python
guards = dict()

guardDayCount = dict()

idx = 0
activeGuard = 0
while idx < len(sorted_list):
    comm = sorted_list[idx]

    if (comm[1][0] == '#'):
        guardID = int(comm[1][1:])
        # print(f'guardID {guardID} is activeguard')

        if guardID not in guards.keys():
            guards[guardID] = [0]*60
            guardDayCount[guardID] = 0
        
        # print(f'guardID {guardID} currently: {guards[guardID]}')
        activeGuard = guardID
        guardDayCount[guardID] +=1
        
        idx +=1
    
    else:
        if(comm[1] == 'asleep'):
            # print(f'Guard: {activeGuard}')

            t = int(str(comm[0])[-4:])
            nextComm = sorted_list[idx+1]
            tNext = int(str(nextComm[0])[-4:])
            guards[activeGuard][t:tNext] = [i+1 for i in guards[activeGuard][t:tNext]]

            # print(f'from {t} to {tNext}')

            idx +=2
```

```python
# Problem 1:

maxSleep = guards[max(guards, key=lambda x: sum(guards[x]))].index((max(guards[max(guards, key=lambda x: sum(guards[x]))] ))) * max(guards, key=lambda x: sum(guards[x]))

print(f'problem 1: {maxSleep}')

```

```python
# Problem 2:
maxRatio = 0
maxValue = 0

for g in guards.keys():
    if max([t / guardDayCount[g] for t in guards[g]]) > maxRatio:
        maxValue = [t / guardDayCount[g] for t in guards[g]].index(max([t / guardDayCount[g] for t in guards[g]])) * g
        maxRatio = max([t / guardDayCount[g] for t in guards[g]])

print(f'Problem 2 solution: {maxValue}')
```
